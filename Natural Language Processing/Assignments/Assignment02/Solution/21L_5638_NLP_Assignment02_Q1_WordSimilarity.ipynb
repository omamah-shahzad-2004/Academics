```python
# Examples of search strings
s1 = 'Bawag bank' # other options: 'Bawag bank', 'Erste', 'Raiffaisen bank'
```

```python
preprocessed_banks = [bank.lower() for bank in banks]
preprocessed_s1 = s1.lower()
```

```python
from difflib import SequenceMatcher
res = []
i = 0
for token in preprocessed_banks:
  res.append([s1, banks[i], SequenceMatcher(None, preprocessed_s1, token).ratio()])
  i = i + 1

df2 = pd.DataFrame(res, columns=['Bank 1', 'Bank 2', 'Score'])
# The outcome is not great, for this search query 'BAWAG Group AG' should have highest similarity
df2.sort_values(by=['Score'], ascending=[False]).head()
```

```python
 #The desired combination has a low score
idx = df2['Bank 2'].isin(['BAWAG Group AG'])

df2[idx].sort_values(by=['Score'], ascending=[False]).head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

#vectorize the texts
tf_idf_matrix = tfidf_vectorizer.fit_transform(preprocessed_banks + [preprocessed_s1])
```

```python
tfidf_vectorizer.vocabulary_
```

```python
print(tf_idf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim = cosine_similarity(tf_idf_matrix[-1], tf_idf_matrix[:-1])
```

```python
res = pd.DataFrame({
    'Bank 1': [s1] * len(banks),
    'Bank 2': banks,
    'Score': cos_sim.flatten()
})
```

```python
df2 = pd.DataFrame(res, columns=['Bank 1', 'Bank 2', 'Score'])
# The outcome is great, for this search query 'BAWAG Group AG' have the highest similarity
df2.sort_values(by=['Score'], ascending=[False]).head()
```
